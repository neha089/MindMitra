```python
import os
import torch
import json
import gradio as gr
from transformers import AutoModelForCausalLM, AutoTokenizer

# ✅ Model Setup
model_name = "Nensee/Llama-2-7b-chat-finetune"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
```

```python
# import os
# import torch
# import json
# import gradio as gr
import tempfile
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from gtts import gTTS

# # ✅ Model Setup
# model_name = "Nensee/Llama-2-7b-chat-finetune"
# tokenizer = AutoTokenizer.from_pretrained(model_name)
# model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")

# tokenizer.pad_token = tokenizer.eos_token if tokenizer.pad_token is None else tokenizer.pad_token

# ✅ Whisper ASR Model (for Speech-to-Text)
ast_model = pipeline("automatic-speech-recognition", model="openai/whisper-small")

# ✅ History File Path
HISTORY_FILE = "chat_history.json"

def load_chat_history():
    if os.path.exists(HISTORY_FILE):
        try:
            with open(HISTORY_FILE, "r") as f:
                data = json.load(f)
                return data if isinstance(data, list) else []
        except json.JSONDecodeError:
            return []
    return []

def save_chat_history(history):
    with open(HISTORY_FILE, "w") as f:
        json.dump(history, f, indent=4)

def clean_response(response):
    response = response.split("User:")[0].strip()
    sentences = response.split(". ")
    return ". ".join(sentences[:-1]) + "." if len(sentences) > 1 else response

def update_user_message(user_input, chat_history):
    chat_history.append({"role": "user", "content": user_input})
    save_chat_history(chat_history)
    return chat_history, ""

def generate_response(chat_history):
    formatted_messages = "\n".join([f"{msg['role'].capitalize()}: {msg['content']}" for msg in chat_history]) + "\nAssistant:"
    inputs = tokenizer(formatted_messages, return_tensors="pt", truncation=True, max_length=2048).to("cuda")
    torch.cuda.empty_cache()

    model.eval()
    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens=100,
            temperature=0.7,
            top_k=50,
            top_p=0.9,
            do_sample=True,
            repetition_penalty=1.1,
            pad_token_id=tokenizer.eos_token_id
        )
    decoded = tokenizer.decode(output[:, inputs["input_ids"].shape[-1]:][0], skip_special_tokens=True)
    cleaned_response = clean_response(decoded)

    chat_history.append({"role": "assistant", "content": cleaned_response})
    save_chat_history(chat_history)
    return chat_history, text_to_speech(cleaned_response)

def transcribe_audio(audio):
    transcript = ast_model(audio)["text"]
    return transcript

def text_to_speech(text):
    tts = gTTS(text)
    temp_audio = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3").name
    tts.save(temp_audio)
    return temp_audio

with gr.Blocks() as demo:
    chatbot = gr.Chatbot(type="messages")
    mode_selector = gr.Radio(["Text Chat", "Voice Chat"], label="Select Chat Mode", value="Text Chat")
    user_input = gr.Textbox(placeholder="Type your message here...")
    voice_input = gr.Audio(sources=["microphone"], type="filepath", interactive=True)
    send_btn = gr.Button("\U000027A1 Send")
    clear_btn = gr.Button("Clear Chat")
    voice_output = gr.Audio()
    history_state = gr.State(load_chat_history())

    def clear_history():
        save_chat_history([])
        history_state.value = []
        return [], "", None

    demo.load(lambda: history_state.value, [], chatbot)

    def handle_chat(mode, text, audio, chat_history):
        if mode == "Text Chat":
            return update_user_message(text, chat_history)
        elif mode == "Voice Chat":
            transcribed_text = transcribe_audio(audio)
            return update_user_message(transcribed_text, chat_history)

    mode_selector.change(lambda mode: (gr.update(visible=(mode == "Text Chat")), gr.update(visible=(mode == "Voice Chat"))), mode_selector, [user_input, voice_input])

    user_input.submit(handle_chat, [mode_selector, user_input, voice_input, history_state], [chatbot, user_input])\
              .then(generate_response, history_state, [chatbot, voice_output])

    send_btn.click(handle_chat, [mode_selector, user_input, voice_input, history_state], [chatbot, user_input])\
            .then(generate_response, history_state, [chatbot, voice_output])

    voice_input.change(handle_chat, [mode_selector, user_input, voice_input, history_state], [chatbot, user_input])\
                .then(generate_response, history_state, [chatbot, voice_output])

    clear_btn.click(clear_history, [], [chatbot, user_input, voice_output])

demo.queue().launch(debug=True, share=True)
```
